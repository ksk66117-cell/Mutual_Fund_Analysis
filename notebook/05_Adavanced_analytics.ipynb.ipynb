```python
import pandas as pd

nav = pd.read_csv("data/raw/02_nav_history.csv")
trans = pd.read_csv("data/raw/08_investor_transactions.csv")
portfolio = pd.read_csv("data/raw/09_portfolio_holdings.csv")

print(nav.head())
print(trans.head())
print(portfolio.head())
```

```python
nav["date"] = pd.to_datetime(nav["date"])

nav["daily_return"] = nav.groupby("amfi_code")["nav"].pct_change()

nav.head()
```

```python
nav["date"] = pd.to_datetime(nav["date"])

nav = nav.sort_values(["amfi_code", "date"])

nav["daily_return"] = nav.groupby("amfi_code")["nav"].pct_change()

nav.head()
```

```python
import numpy as np

risk_report = nav.groupby("amfi_code")["daily_return"].apply(
    lambda x: pd.Series({
        "VaR_95": np.percentile(x.dropna(), 5),
        "CVaR_95": x[x <= np.percentile(x.dropna(), 5)].mean()
    })
).unstack()

risk_report.head()
```

```python
risk_report.to_csv("var_cvar_report.csv")
```

```python

fund = nav[nav["amfi_code"] == nav["amfi_code"].iloc[0]].copy()

fund["rolling_sharpe"] = (
    fund["daily_return"].rolling(90).mean()
    / fund["daily_return"].rolling(90).std()
) * np.sqrt(252)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(fund["date"], fund["rolling_sharpe"])

plt.title("90-Day Rolling Sharpe Ratio")
plt.xlabel("Date")
plt.ylabel("Sharpe Ratio")
plt.grid(True)

plt.savefig("rolling_sharpe_chart.png")

plt.show()
```

```python
fund[["date", "rolling_sharpe"]].to_csv(
    "rolling_sharpe.csv",
    index=False
)
```

```python
age_group = trans["age_group"].value_counts()

import matplotlib.pyplot as plt

plt.figure(figsize=(7,7))
plt.pie(age_group,
        labels=age_group.index,
        autopct="%1.1f%%")

plt.title("Investor Age Group Distribution")

plt.savefig("cohort_chart.png")

plt.show()
```

```python
sip = trans[trans["transaction_type"] == "SIP"]

sip_monthly = sip.groupby(
    pd.to_datetime(sip["transaction_date"]).dt.to_period("M")
).size()

plt.figure(figsize=(12,5))
sip_monthly.plot()

plt.title("Monthly SIP Transactions")
plt.xlabel("Month")
plt.ylabel("Number of SIP Transactions")
plt.grid(True)

plt.savefig("sip_continuity.png")

plt.show()
```

```python
def recommend_fund(risk):

    if risk == "Low":
        return "Debt Fund"

    elif risk == "Moderate":
        return "Hybrid Fund"

    elif risk == "High":
        return "Equity Fund"

print(recommend_fund("High"))
```
